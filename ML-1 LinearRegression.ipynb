Linear Regression

* using house pricing dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston #house pricing dataset is present in sklearn dataset by default
```

```python
df= load_boston()
```

```python
type(df)
```

```python
df  #data is on form of key value pair 
```

```python
#converting df into dataframe

data = pd.DataFrame(df.data)
```

```python
#adding target dataser
#Pandas DataFrame.columns attribute return the column labels of the given Dataframe

#take feature_names which is araay of features and their values and assign to data.columns

data.columns = df.feature_names #in df, fetch feature_names section
data.head()


#these are independant features
```

```python
#create new feature Price which is dependant

data["Price"]= df.target
data.head()
```

```python
#dividing the dataset into independant and dependant features
# X independant and y is dependant 

X= data.iloc[:, :-1] #skip last feature. first : means all row, all col: except -1
y= data.iloc[:, -1] # all rowsm -1 means last col
```

```python
X.head()
```

```python
y.head()
```

```python
#Applying Linear Regression

from sklearn.linear_model import LinearRegression

lin_reg= LinearRegression()
```

```python
#cross validation: we will divide train and test data in every combination taken by model 
#and whose accuracy is better we wll combine
# suppose there are 100 columns. and crossval is 5. in 1st iteration we will take first 20 as train and 
#rest as test. 2nd iter- 2nd 20 column as train and 1st 20 and from 40 till 100 as test. similarly 5 times

from sklearn.model_selection import cross_val_score
mse= cross_val_score(lin_reg, X, y, scoring='neg_mean_squared_error', cv=5) 
#mse= mean squared error
# we can also split and provide X_train and y_train (preffered)

```

```python
# just to check mse: we are doin  5 cross val so we get 5 
print(mse)
```

```python
mean_mse=np.mean(mse)
print(mean_mse)
```
